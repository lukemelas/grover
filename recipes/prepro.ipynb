```python
import pandas as pd
import json
```

```python
data = pd.read_csv('879-recipes-sam-sifton-cleaned.csv', index_col=0)
```

```python
data.head(3)
# data['Author'].value_counts().head()
```

```python
fname = '879-recipes-prepro.jsonl'
with open(fname, 'w') as f:

    for i, x in data.iterrows():

        output = json.dumps({
            "title": x['Name'],
            "text": "",
            "authors": ["Sam Sifton"],
            "publish_date":"04-12-2019",
            "status":"success",
            "warc_date": "2019-04-25T16:22:51Z",
            "split":"gen",
            "inst_index": i,
            "domain": "nytimes.com",
            "summary": "",
        })

        print(output, file=f)
```
